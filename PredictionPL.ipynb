```python
matches = matches.sort_values(["season", "team", "date"]).reset_index(drop=True)
```

```python
matches["result_code"] = matches["result"].map({
    "L":0,
    "W":3,
    "D":1
})
```

```python
matches["form_5"] = matches.groupby(["season", "team"])["result_code"].transform(lambda x: x.shift().rolling(5, min_periods=1, closed="left").mean())
```

```python
matches["form_5"] = matches["form_5"].fillna(1)
```

```python
matches["avg_gf_5"] = matches.groupby("team")["gf"].transform(lambda x: x.shift().rolling(5, min_periods=1, closed="left").mean())
```

```python
matches["avg_ga_5"] = matches.groupby("team")["ga"].transform(lambda x: x.shift().rolling(5, min_periods=1, closed="left").mean())
```

```python
matches["avg_gd_5"] = matches["avg_gf_5"] - matches["avg_ga_5"]
```

```python
matches[["avg_gd_5","avg_ga_5","avg_gf_5"]] = matches[["avg_gd_5","avg_ga_5","avg_gf_5"]].fillna(0)
```

```python
matches["roll_xg_10"] = matches.groupby("team")["xg"].transform(lambda x: x.shift().rolling(10, min_periods=1, closed="left").mean())
```

```python
matches["roll_xg_10"] = matches["roll_xg_10"].fillna(0)
```

```python
matches["opp_result_code"] = matches["result"].map({
    "W":0,
    "D":1,
    "L":3
})
```

```python
matches["opp_form_5"] = matches.groupby("opp_code")["opp_result_code"].transform(lambda x: x.rolling(5, min_periods=1, closed="left").mean())
```

```python
matches["opp_form_5"] = matches["opp_form_5"].fillna(1)
```

```python
matches["team_form_5"] = matches["form_5"]
```

```python
matches = matches.drop(columns="form_5")
```

```python
matches["form_difference_5"] = matches["team_form_5"] - matches["opp_form_5"]
```

```python
matches["avg_sh_5"] = matches.groupby(["season","team"])["sh"].transform(lambda x: x.shift().rolling(5, min_periods=1).mean())
```

```python
matches["avg_sot_5"] = matches.groupby(["season","team"])["sot"].transform(lambda x: x.shift().rolling(5, min_periods=1).mean())
```

```python
matches[["avg_sh_5","avg_sot_5"]] = matches[["avg_sh_5","avg_sot_5"]].fillna(0)
```

```python
matches["total_team_points"] = (matches.groupby(["season", "team"])["result_code"].cumsum().shift(fill_value=0))
```

```python
train = matches[matches["date"] < '2024-08-17']
test = matches[matches["date"] >= '2024-08-17']
```

```python
predictors = ["venue_code", "opp_code", "hours", "day_code", "team_form_5", 
              "avg_gf_5", "avg_ga_5", "avg_gd_5", "roll_xg_10", "opp_form_5", 
              "form_difference_5", "avg_sh_5", "avg_sot_5", "total_team_points"]
```

```python
def prediction_model(matches, predictors, train, test):

    rf = RandomForestClassifier(n_estimators=200, min_samples_split=10, class_weight="balanced", max_depth=12, random_state=1)

    rf.fit(train[predictors], train["target"])

    preds = rf.predict(test[predictors])

    print("Train size:", len(train))
    print("Test size:", len(test))

    acc = accuracy_score(test["target"], preds)

    combine = pd.DataFrame({
    'actual':test['target'], 
    'prediction':preds
    })

    return acc, preds, combine
    
```

```python
acc, preds, combine = prediction_model(matches, predictors, train, test)
```
