```python
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
```

```python
matches["hours"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
#matches["target"] = (matches[result] == "W")
matches["target"] = matches["result"].map({
    "L": 0,
    "D": 1, 
    "W": 2
})
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=100, min_samples_split=10, random_state=1)
```

```python
train = matches[matches["date"] < '2024-08-17']
```

```python
test = matches[matches["date"] >= '2024-08-17']
```

```python
predictors = ["venue_code", "opp_code", "hours", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["target"], preds)
```

```python
acc
```

```python
combine = pd.DataFrame({
    'actual':test['target'], 
    'prediction':preds
})
```

```python
pd.crosstab(index=combine["actual"], columns=combine["prediction"])
#    "L" is 0
#    "D" is 1
#    "W" is 2
```

```python
from sklearn.metrics import precision_score
```

```python
precision_score(test["target"], preds, average="weighted")
```

```python
matches["result_code"] = matches["result"].map({
    "L":0,
    "W":3,
    "D":1
})
```

```python
matches["form_5"] = matches.groupby("team")["result_code"].transform(lambda x: x.rolling(5, min_periods=1, closed="left").mean())
```

```python
matches["form_5"] = matches["form_5"].fillna(0)
```

```python
matches["avg_gf_5"] = matches.groupby("team")["gf"].transform(lambda x: x.rolling(5, min_periods=1, closed="left").mean())
```

```python
matches["avg_ga_5"] = matches.groupby("team")["ga"].transform(lambda x: x.rolling(5, min_periods=1, closed="left").mean())
```

```python
matches["avg_gd_5"] = matches["avg_gf_5"] - matches["avg_ga_5"]
```

```python
matches[["avg_gd_5","avg_ga_5","avg_gf_5"]] = matches[["avg_gd_5","avg_ga_5","avg_gf_5"]].fillna(0)
```

```python
matches["roll_xg_10"] = matches.groupby("team")["xg"].transform(lambda x: x.rolling(10, min_periods=1, closed="left").mean())
```

```python
matches["roll_xg_10"] = matches["roll_xg_10"].fillna(0)
```

```python
matches["roll_xga_10"] = matches.groupby("team")["xga"].transform(lambda x: x.rolling(10, min_periods=1, closed="left").mean())
```

```python
matches["roll_xga_10"] = matches["roll_xga_10"].fillna(0)
```

```python
matches = matches.drop(columns="roll_xg_5")
```

```python
predictors = ["venue_code", "opp_code", "hours", "day_code", "form_5", "avg_gf_5", "avg_ga_5", "avg_gd_5", "roll_xg_10"]
```

```python
def prediction_model(matches, predictors):
    train = matches[matches["date"] < '2024-08-17']
    test = matches[matches["date"] >= '2024-08-17']

    rf = RandomForestClassifier(n_estimators=200, min_samples_split=10, max_depth=9, random_state=1)
            #     max_depth=8,
            # class_weight="balanced"

    rf.fit(train[predictors], train["target"])

    preds = rf.predict(test[predictors])

    print("Train size:", len(train))
    print("Test size:", len(test))

    acc = accuracy_score(test["target"], preds)

    combine = pd.DataFrame({
    'actual':test['target'], 
    'prediction':preds
    })

    return acc, preds, combine
    
```

```python
acc, preds, combine = prediction_model(matches, predictors)
```

```python
acc
```

```python
pd.crosstab(index=combine["actual"], columns=combine["prediction"])
#    "L" is 0
#    "D" is 1
#    "W" is 2
```

```python
precision_score(test["target"], preds, average="weighted")
```
