```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(10, min_periods=1, closed="left").mean()
    group[new_cols] = rolling_stats
    group[new_cols] = group[new_cols].fillna(0)
    return group
```

```python
cols = ["xg", "gf", "ga", "sh", "sot", "dist", "fk", "pk"]
```

```python
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling = matches_rolling.droplevel('team')
```

```python
matches_rolling
```

```python
matches_rolling.shape
```

```python
matches_rolling
```

```python
matches_rolling["result_code"] = matches_rolling["result"].map({
    "L":0,
    "W":3,
    "D":1
})
```

```python
matches_rolling["team_form_5"] = matches_rolling.groupby(["season", "team"])["result_code"].transform(lambda x: x.rolling(5, min_periods=1, closed="left").mean())
```

```python
matches_rolling["team_form_5"] = matches_rolling["team_form_5"].fillna(1)
```

```python
matches_rolling.shape
```

```python
# Create lookup using the team as the main team (not as opponent)
team_form_lookup = matches_rolling[["season", "team", "date", "team_form_5"]].copy()

# Merge opponent's form by matching opponent name with team name
matches_rolling = matches_rolling.merge(
    team_form_lookup.rename(columns={"team": "opponent", "team_form_5": "opp_form_5"}),
    on=["season", "opponent", "date"],
    how="left"
)
```

```python
matches_rolling["form_difference_5"] = matches_rolling["team_form_5"] - matches_rolling["opp_form_5"]
```

```python
matches_rolling["total_team_points"] = (matches_rolling.groupby(["season", "team"])["result_code"].cumsum().shift(fill_value=0))
```

```python
matches_rolling.columns
```

```python
matches_rolling = matches_rolling.drop(columns=["attendance", "notes" ])
```

```python
predictors = ["venue_code", "opp_code", "hours", "day_code",
              "xg_rolling", "gf_rolling", "ga_rolling", "sh_rolling",
              "sot_rolling", "dist_rolling", "fk_rolling", "pk_rolling",
              "team_form_5", "total_team_points", "opp_form_5", "form_difference_5" ]

#  
```

```python
def prediction_model(matches, predictors, train, test):

    rf = RandomForestClassifier(n_estimators=200, min_samples_split=10, class_weight="balanced", max_depth=18, random_state=1)

    rf.fit(train[predictors], train["target"])

    preds = rf.predict(test[predictors])

    print("Train size:", len(train))
    print("Test size:", len(test))

    acc = accuracy_score(test["target"], preds)
    error = precision_score(test["target"], preds, average='weighted')

    combine = pd.DataFrame({
    'actual':test['target'], 
    'prediction':preds
    })

    return acc, preds, combine, error
    
```

```python
acc, preds, combine, error = prediction_model(matches_rolling, predictors, train, test)
```
